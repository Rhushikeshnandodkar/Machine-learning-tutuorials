```python
import pandas as pd
import random
```

```python
df = pd.read_csv('Top Indian Places to Visit.csv')
```

```python
df.head()
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data_list = []

# Loop through the 'Name' column and months
for i in df['Unnamed: 0']:
    for j in months:
        # Append a dictionary with the data
        data_list.append({
            'month': j,
            'Place': i,
            'count': random.randint(1, 4)
        })
random_df = pd.DataFrame(data_list)
```

```python
random_df
```

```python
random_df.to_csv('crowdid.csv', index=True)
```
