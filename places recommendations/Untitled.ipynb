```python
df['Type'] = df['Type'].apply(lambda x : str(x).split())
df['Best Time to visit'] = df['Best Time to visit'].apply(lambda x : str(x).split())

df['fee_category'] = df['fee_category'].apply(lambda x : str(x).split())
df['establishment_category'] = df['establishment_category'].apply(lambda x : str(x).split())
df['Type'] = df['Type'].apply(lambda x : str(x).split())

df['Type']
```

```python
 df["Type"] = df['Type'].apply(lambda x: [i.replace(" ", "") for i in x])
df["Type"][3]
```

```python
# df['tags'] = df["Zone"] + df['City'] + df['Type'] 
df['City'] = df['City'].apply(lambda x : str(x).split())
df['State'] = df['State'].apply(lambda x : str(x).split())
df['Zone'] = df['Zone'].apply(lambda x : str(x).split())
df['DSLR Allowed'] = df['DSLR Allowed'].apply(lambda x : str(x).split())
```

```python
df['Significance'] =  df['Significance'].apply(lambda x : str(x).split())
```

```python
df['Tags'] = df['Zone'] + df['Type'] + df['Significance'] + df['State']
```

```python
new_df = df[['Unnamed: 0', 'Name', 'Tags']]
```

```python
new_df
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x : " ".join(x))
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cv.get_feature_names_out()
```

```python
def recommend(place):
    place_index = new_df[new_df['Name'] == place].index[0]
    distances = similarity[place_index]
    place_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:16]
    for i in place_list:
        print(new_df.iloc[i[0]].Name)
```

```python
recommend("Jagannath Temple")
```
